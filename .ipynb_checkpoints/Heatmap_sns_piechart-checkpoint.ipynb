```python
#import matplot libarary
import matplotlib.pyplot as plt

#import seaborn library
import seaborn as sns
```

```python
#to show plot on notebook
%matplotlib inline
```

```python
#load flights data from sns dataset(built in)
filght_data =sns.load_dataset('flights')
```

```python
filght_data
```

```python
#view top 5 records
filght_data.head()
```

```python
#use pivot method to rearrage the datset
filght_data = filght_data.pivot('month','year','passengers')
filght_data
```

```python
#view datset
filght_data.head()
```

```python
#use heatmap method to generate the heatmap of the flights data
sns.heatmap(filght_data)
```

## piechart

```python
#job data in percentile
job_data =['40','20','30','60','45']

#define labels as different departments
labels ='IT','FINANCE','MARKETING','BANKING','HR'
#EXPLODE THE 1ST SLICE WHICH IS IT
explode =(0.05,0.05,0,0,0)
#draw the pie chart and set the parameters
plt.pie(job_data,labels =labels,explode=explode)
#show the plot
plt.show()
```
