```python
# import all the dependencies
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib notebook
import numpy as np
```

```python
# Read the csv file
cust_df = pd.read_csv("company_sales_data.csv")
```

```python
cust_df.head()
```

```python
profitList =cust_df['total_profit'].tolist()
monthList  =cust_df['month_number'].tolist()

plt.plot(monthList, profitList, label = 'Profit data of last year', 
      color='r', marker='o', markerfacecolor='k', 
      linestyle='--', linewidth=3)
      
plt.xlabel('Month Number')
plt.ylabel('Profit in dollar')
plt.legend(loc='lower right')
plt.title('Company Sales data of last year')
plt.xticks(monthList)
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

```python
profitlist = cust_df['total_profit'].tolist()
monthlist = cust_df['month_number'].tolist()

plt.plot(monthlist,profitlist,label ='profit data of last year',
         color ='r', marker ='o',markerfacecolor ='k',
         linestyle ='--',linewidth =3)
plt.xlabel('Month number')
plt.ylabel(" profit in dollar")
plt.legend(loc = 'lower right')
plt.title(" Comany sales data of last year")
plt.xticks(monthList)

plt.yticks([100000,200000,300000,400000,500000])
plt.show()
```

```python
total_units = cust_df['total_units'].tolist()
monthlist = cust_df['month_number'].tolist()
facecream = cust_df['facecream'].tolist()
facewash = cust_df['facewash'].tolist()
toothpaste= cust_df["toothpaste"].tolist()
bathingsoap = cust_df['bathingsoap'].tolist()
shampoo = cust_df['shampoo'].tolist()
moisturizer =cust_df['moisturizer'].tolist()
plt.plot(monthlist ,facecream,label =" face cream sales data",color = 'r',marker ='o',linewidth =3)
plt.plot(monthlist ,facewash,label =" facewash sales data",color = 'blue',marker ='*',linewidth =3)
plt.plot(monthlist ,toothpaste,label ="toothpaste sales data",color = 'green',marker ='^',linewidth =3)
plt.plot(monthlist ,bathingsoap,label =" bathingsoap data",color = 'r',marker ='o',linewidth =3)
plt.legend(loc ='upperleft')
```

```python
total_units = cust_df['total_units'].tolist()
monthlist = cust_df['month_number'].tolist()
facecream = cust_df['facecream'].tolist()
plt.scatter(monthlist,facecream,label = 'sales data')

plt.grid(True,linewidth= 1,linestyle ='-')
plt.show()
```

```python
total_units = cust_df['total_units'].tolist()
monthlist = cust_df['month_number'].tolist()
facecream = cust_df['facecream'].tolist()
facewash = cust_df['facewash'].tolist()

plt.bar([a-0.25 for a in monthlist],facecream ,width= 0.25,label ='face cream data',align ='edge')
plt.bar([a+0.25 for a in monthlist],facewash,width =0.25,label ='facewash data',align ='edge')
plt.grid(True,linewidth= 1,linestyle ='-')
plt.show()
plt.savefig("facewash.png",dip=150)
```

```python
total_profit = cust_df['total_profit'].tolist()
plt.hist(total_profit)
plt.show()
```

```python
monthlist = cust_df['month_number'].tolist()
labels =['Facecream','Facewash','Toothpaste','Shampoo']
salesdata = [cust_df['facecream'].sum(),cust_df['facewash'].sum(),cust_df['toothpaste'].sum(),cust_df['shampoo'].sum()]
plt.axis("equal")
plt.pie(salesdata,labels= labels, autopct ="%1.1f%%")
plt.legend(loc ='lower right')
plt.title('salesdata')
plt.show()

```

```python
monthlist  =cust_df['month_number'].tolist()
bathingsoap   = cust_df['bathingsoap'].tolist()
faceWashSalesData   = cust_df['facewash'].tolist()
f, axarr =plt.subplots(2,sharex =True)
axarr[0].plot(monthlist,bathingsoap,label ='bathingsoap sales data',color = 'r',linewidth=3)
axarr[0].set_title('sales data of bathsoap')

axarr[1].plot(monthlist,faceWashSalesData ,label ='facewash sales data',color='green',linewidth =4)

axarr[1].set_title('data')
plt.xticks(monthlist)
plt.xlabel('month number')
plt.ylabel('sales units in number')
plt.show()
           
```

```python
monthlist  =cust_df['month_number'].tolist()
bathingsoap   = cust_df['bathingsoap'].tolist()
faceWashSalesData   = cust_df['facewash'].tolist()
plt.plot([],[],color = 'm',label = 'bathingsoap ',linewidth =5)
plt.plot([],[],color = 'm',label = 'faceWashSalesData  ',linewidth =5)
plt.stackplot(monthlist,bathingsoap,faceWashSalesData,colors=['m','r','c'])
plt.show()
```
