## Line properties
Line properies:1.Alpha
2.Animated

Plot Graphics:
1.linestyle
2.line width
3.marker style


```python
#import the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt
from matplotlib import style
%matplotlib inline
```

```python
#website traffice data
#number of users/visitors on the website
web_customers =[123,640,44,678,99,80,1900,1295,]
#Time distribution(hourly)
time_hrs = [7,8,9,12,14,15,16,18]
```

```python
#select the style of the plot
style.use('ggplot')
#plot the web site traffif data (X_axis hrs and y axis as number of users)
plt.plot(time_hrs,web_customers,color ='b',linestyle ='--',linewidth =2.5)
#set the title of the plot
plt.title('web site traffic')
#set the label for x axis
plt.xlabel('Hrs')
# set the X axis and y axis
plt.axis([6.5,17.5,50,2000])

#set label for y axis
plt.ylabel('Number of users')
plt.legend()
plt.show()
```

```python
#select the style of the plot
style.use('ggplot')
#plot the web site traffif data (X_axis hrs and y axis as number of users)
#also setting the alpha value for transparency
plt.plot(time_hrs,web_customers,label ='Webtraffic',alpha=.5)
#set the title of the plot
plt.title('web site traffic')
#set the label for x axis
plt.xlabel('Hrs')
# set the X axis and y axis
plt.axis([6.5,17.5,50,2000])

#Annotate
plt.annotate('Max',ha='center',va ='bottom',xytext =(14,1900),xy =(16,1950),arrowprops ={'facecolor':'green'})
#set label for y axis
plt.ylabel('Number of users')
plt.legend()
plt.show()
```
