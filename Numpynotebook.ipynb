```python
# write a anogram
str1 ='actd'
str2 ='cat'
from collections import Counter
def check_anagram(s1,s2):
    s1 = s1.lower().replace(" ",' ')
    s2 =s2.lower().replace(" ",' ')
    return Counter(s1)== Counter(s2)
check_anagram(str1,str2)
                 
```

```python
def implement_counter(s1):
    c={}
    for i in s1:
        if i in c:
            c[i]+=1
        else:
            c[i]= 1
    return c
s1="billie"
implement_counter(s1)
```
