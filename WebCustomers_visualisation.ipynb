```python
#import the dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as  plt
from matplotlib import style
%matplotlib inline
```

```python
#website traffice data
#number of users/visitors on the website
web_customers =[123,640,44,678,99,80,3456,1295,]
#Time distribution(hourly)
time_hrs = [7,8,9,12,14,15,16,18]
```

```python
#select the style of the plot
style.use('ggplot')
#plot the web site traffif data (X_axis hrs and y axis as number of users)
plt.plot(time_hrs,web_customers)
#set the title of the plot
plt.title('web site traffic')
#set the label for x axis
plt.xlabel('Hrs')

#set label for y axis
plt.ylabel('Number of users')
plt.show()
```
